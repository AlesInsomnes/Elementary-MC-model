```python
Ax = 5.85E-10
Ay = 1.78E-10
Az = 4.41E-10

# A1x = Ax / (2 ** (1 / 3))
# A1y = Ay / (2 ** (1 / 3))
# A1z = Az / (2 ** (1 / 3))

# A1x = Ax / (7 ** (1 / 3))
# A1y = Ay / (7 ** (1 / 3))
# A1z = Az / (7 ** (1 / 3))

A1x = Ax / (14 ** (1 / 3))
A1y = Ay / (14 ** (1 / 3))
A1z = Az / (14 ** (1 / 3))

g100 = 0.41
g010 = 0.54
g001 = 0.22

Ω = A1x * A1y * A1z

EX = 2 * g100 * A1y * A1z
EY = 2 * g010 * A1z * A1x
EZ = 2 * g001 * A1x * A1y

Eisol = EX + EY + EZ
E3 = EY + EX
E2 = EZ + EY
E1 = EZ + EX

kB = 1.380649e-23
T = 300.0
kT = kB * T
Ceq = np.exp(- Eisol / kT)

print(
    """
Sizes:
    Ax: {:.5e},    Ay: {:.5e},    Az: {:.5e}
   A1x: {:.5e},   A1y: {:.5e},   A1z: {:.5e}

Volume:
   Ω = A1x * A1y * A1z = {:.5e} * {:.5e} * {:.5e} = {:.5e}

Surface tensions:
   g100: {:.5e}, g010: {:.5e}, g001: {:.5e}

Energys:
1)
   EX = 2.0 * g100 * A1y * A1z = 2.0 * {:.5e} * {:.5e} * {:.5e} = {:.5e}
   EY = 2.0 * g010 * A1z * A1x = 2.0 * {:.5e} * {:.5e} * {:.5e} = {:.5e}
   EZ = 2.0 * g001 * A1x * A1y = 2.0 * {:.5e} * {:.5e} * {:.5e} = {:.5e}

2)
   Eisol = EX + EY + EZ = {:.5e} + {:.5e} + {:.5e} = {:.5e}

3)
   E3 = EY + EX = {:.5e} + {:.5e} = {:.5e}
   E2 = EZ + EY = {:.5e} + {:.5e} = {:.5e}
   E1 = EZ + EX = {:.5e} + {:.5e} = {:.5e}

kB = {:.5e}; T = {:.5f};
kT = kB * T = {:.5e} * {:.5f} = {:.5e};
Ceq = exp(- Eisol / kT) = exp(- {:.5e} / {:.5e} ) = {:.5e};
""".format(
        Ax, Ay, Az,
        A1x, A1y, A1z,
        A1x, A1y, A1z,
        Ω,
        g100, g010, g001,
        g100, A1y, A1z, EX,
        g010, A1z, A1x, EY,
        g001, A1x, A1y, EZ,
        EX, EY, EZ, Eisol,
        EY, EX, E3,
        EZ, EY, E2,
        EZ, EX, E1,
        kB, T,
        kB, T, kT,
        Eisol, kT, Ceq,
    ).strip()
)
```
